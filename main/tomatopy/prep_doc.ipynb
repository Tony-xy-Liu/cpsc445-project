```python
import os
import pandas as pd
import numpy as np
from local.caching import load, save

df: pd.DataFrame = load('biocyc_metabolite_usage', alt_workspace="../prep/")
df.head()
```

```python
mat = df.to_numpy()
variance = mat.var(axis=0)
variance.shape
```

```python
mean = mat.mean(axis=0)
mean.shape
```

```python
for name, v, av in sorted(zip(df.columns, np.sqrt(variance), mean), key=lambda t: t[1], reverse=True):
    print(name, '\t', v, '\t', av)
```

```python
for name, row in df.iterrows():
    vals = [(c, v) for c, v in zip(df.columns, row)]
    vals = sorted(vals, key=lambda t: t[1], reverse=True)
    for c, v in vals:
        print(c, v)
    break
```

```python
from tomotopy.utils import Corpus
```

```python
corp = Corpus()

columns = df.columns
label_len = max([len(n) for n in columns])
for i, (_, row) in enumerate(df.iterrows()):
    if i % 16 == 0: print(i, row.sum(), " "*25, end='\r')
    words = np.empty(shape=(row.sum(),), dtype=f'<U{label_len}')
    doc_i = 0
    for j, c in enumerate(row):
        if c == 0: continue
        for _ in range(c):
            words[doc_i] = columns[j]
            doc_i += 1

    corp.add_doc(words)
```

```python
corp.save('./cache/biocyc_metabolite_usage_corpus')
```
