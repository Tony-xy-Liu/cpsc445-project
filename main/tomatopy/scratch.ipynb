```python
import tomotopy as tp
from tomotopy.utils import Corpus
import numpy as np
```

```python
corp = Corpus.load('./cache/biocyc_metabolite_usage_corpus')
K = 100
model = tp.CTModel(k=K, rm_top=35, min_cf=10)
model.add_corpus(corp)
```

```python
for i in range(10):
    model.train(iter=10, workers=12)

model.summary()
```

```python
for i in range(K):
    for lst in model.get_topic_words(i):
        print(lst)
    print()
```

```python
for i in range(400):
    print(i, end='\r')
    model.train(iter=25, workers=12)
model.summary()
```
