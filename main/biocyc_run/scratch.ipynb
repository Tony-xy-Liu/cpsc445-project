```python
import os
from pathlib import Path
import numpy as np
import pandas as pd
from plotly import graph_objects as go, subplots as sp

from local.caching import save, load
from local.figures import layout
from biocyc_facade.pgdb import Pgdb, Dat
from model import MNetwork, Cluster
```

```python
for f in os.listdir("./cache/"):
    if not f.endswith("tree.pkl.gz"): continue

    # print(f)
    # break
    tree = load(f)
    break
```

```python
points = []
def traverse(tree, depth):
    if not isinstance(tree, tuple):
        return 1, tree # leaf
    
    a, b, q, gq = tree

    na, grpa = traverse(a, depth+1)
    nb, grpb = traverse(b, depth+1)

    count, nodes = na+nb, grpa+grpb
    points.append((depth, count, q, gq))
    return count, nodes

count, nodes = traverse(tree, 0)
count
```

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x = [tup[1] for tup in points],
        y = [tup[3] for tup in points],
        mode="markers",
        marker=dict(size=5),
        showlegend=False,
    ),
)

fig.update_annotations(font_size=24)
fig.update_layout(go.Layout(layout))

fig.show()
```
